```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score 
```

```python
%matplotlib inline
```

## <center>                            $R^2$ = $1 - \frac{ \sum (y-\hat y)^2}{∑(y- \bar y)^2} $

```python
def compute_r2_numpy(y_true, y_predicted):
    variance_model = np.sum(np.square(y_true - y_predicted))
    variance_données = np.sum(np.square(y_true - np.mean(y_true)))
    R_squared = 1 - (variance_model/variance_données)
    return R_squared
     
```

```python
x = np.arange(4)
y_true = np.arange(4)
```

```python
plt.scatter (x, y_true)
plt.show()
```

```python
y_pred = np.array([0, 1, 2, 3])
```

```python
plt.scatter (x, y_true, s=50, color='red')
plt.plot(x, y_pred)
plt.show()
```

```python
print( "R2 calculé :",compute_r2_numpy(y_true, y_pred) )
print( "R2 de scikit learn :", r2_score(y_true, y_pred))
```
